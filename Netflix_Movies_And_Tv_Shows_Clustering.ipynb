```python
# With respect to each rating assigning it into group of categories

rating_age = {"TV-PG" : "Older Kids",
              "TV-MA" : "Adults",
              "TV-Y7-FV" : "Older Kids",
              "TV-Y7" : "Older Kids",
              "TV-14" : "Teens",
              "R" : "Adults",
              "TV-Y" : "Kids",
              "NR" : "Adults",
              "PG-13" : "Teens",
              "TV-G" : "Kids",
              "PG" : "Older Kids",
              "G" : "Kids",
              "UR" : "Adults",
              "NC-17" : "Adults"
              }

df_hypo['Target_ages'] = df_hypo['rating'].replace(rating_age)
df_hypo['Target_ages'].unique()
```

```python
df_hypo['duration'] = df_hypo['duration'].str.extract("(\d+)")
df_hypo['duration'] = pd.to_numeric(df_hypo['duration'])
```

```python
# Groupby target & duration_edges

group_by = df_hypo[["duration" , 'Target_ages']].groupby(by = 'Target_ages')
group = group_by.mean().reset_index()

group
```

```python
# In A and B variable grouping values

A = group_by.get_group('Kids')
B = group_by.get_group("Older Kids")
```

```python
M1 = A.mean()
S1 = A.std()
```

```python
M2 = B.mean()
S2 = B.std()
```

```python
print("Mean for Movie rated for kids {} \n Mean for movie rated for Older Kids {}".format(M1,M2))
print("Std Movie rated for kids {} \n Std for movie rated for Older Kids {}".format(S1,S2))
```

```python
# Import Stats

from scipy import stats

# Length of groups

n1 = len(A)
n2 = len(B)
print(n1,n2)
```

```python
dof = n1 + n2 - 2
print("dof", dof)
```

```python
sp_2 = ((n2-1)* S1**2 + (n1-1) * S2**2)/dof
print("SP_2", sp_2)
```

```python
sp = np.sqrt(sp_2)
print("SP", sp)
```

```python
# T-Value

t_val = (M1-M2)/(sp *np.sqrt(1/n1 + 1/n2))
print("T_Value",t_val[0])
```

```python
# t-distribution

stats.t.ppf(0.025, dof)
```

```python
# t-distribution

stats.t.ppf(0.975, dof)
```

* The null hypothesis stating that movies rated for kids and older kids on Netflix are at least two hours long is rejected based on the t-value falling outside the specificed range. Therefore, it can be concluded that movies targeted for kids and older kids on Netflix have a duration that is less than 2 hours.

```python
def remove_punction(text):
  import string
# Replace the punctuation with no space
# Which is effect delete the panctuation marks
  translator = str.maketrans("","", string.punctuation)
# Return the text stripped of panctuation marks
  return text.translate(translator)


# Applying above function on text features

df['description'] = df['description'].apply(remove_punction)
```

```python
import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords

sw = stopwords.words('english')
```

```python
# Defining Stopwords

def stopwords(text):
  # It is a function to remove stopwords
  text = [word for word in text.split() if word not in sw]
  # Joining the WT words with space seperator
  return " ".join(text)


# Applying above function on text function
df['description'] = df['description'].apply(stopwords)
df['description'][0]
```

#**Clustering Algorithms**

###**1) K-Means**

We know that K-means requires elbow method to know number of cluster.

```python
# Finding optimal number of clustering using the ellbow method\

from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,30):

  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,30), wcss)
plt.title("The Elbow Method Graphic")
plt.xlabel("Number of cluster (K)")
plt.ylabel("WCSS_list")
plt.show()
```

```python
from sklearn.metrics.cluster import silhouette_samples
from sklearn.metrics import silhouette_score
# Silhoute score of cluster
sill = []
for i in range(2,30):
  model = KMeans(n_clusters=i, init='k-means++', random_state=51)
  model.fit(x)

  y1 = model.predict(x)
  score = silhouette_score(x, y1)
  sill.append(score)
  print("Cluster: %d \t Silhoute : %0.4f" %(i, score))
```

```python
# Ploting silhoite's score

plt.figure(figsize = (14,6))
plt.plot(sill, 'bs--')
plt.xticks(list(range(2,31)), list(range(2,31)))
plt.xlabel("Number of cluster")
plt.show()
```

* Based on the elbow method and silhouette score analysis, it is suggested to form 26 clusters for the given dataset. The elbow method helps determine the optimal number of clusters by evaluating the distortion or inertia, while the silhouette score measures the quality and separation of the clusters. Both methods indicate that 26 clusters would be appropriate for the dataset based on the available information.italicized text

```python
# Traning the k-means model on a dataset

Kmeans = KMeans(n_clusters = 26, init = 'k-means++', random_state=42)
y_pred = Kmeans.fit_predict(x)
```
